```python
import pandas as pd
import sqlite3
df = pd.read_csv("cleaned_global_ecommerce_sales.csv")
df.head()
```

```python
conn = sqlite3.connect("ecommerce.db")
```

```python
df.to_sql("sales", conn, if_exists="replace", index=False)
```

Display All Records

```python
query = "SELECT * FROM sales;"

result = pd.read_sql(query, conn)
result.head()
```

Filter Records

```python
query = """
SELECT *
FROM sales
WHERE total_sales > 500;
"""

result = pd.read_sql(query, conn)
result.head()
```

Category-wise Sales

```python
query = """
SELECT product_category,
SUM(total_sales) AS total_sales
FROM sales
GROUP BY product_category;
"""

result = pd.read_sql(query, conn)
result
```

Use a Parameterized Query

```python
sales_limit = 500

query = """
SELECT *
FROM sales
WHERE total_sales > ?;
"""

result = pd.read_sql(query, conn, params=[sales_limit])
result.head()
```

Save SQL Results to a CSV File

```python
result.to_csv("category_sales.csv", index=False)
```

This creates a file named category_sales.csv

```python
conn.close()
```
